### Imported Library needed

```python
# Add the library usedn
import pandas as pd
import numpy as np
```

```python
# Enter a dataset that will be used as a test material
data = pd.read_excel("NMC1-9.xlsx")
```

Make a calculation function that will be used in testing.
Tests carried out namely to search
1. Obstacles in Ohm Units (R)
    where r = v/i
2. Power in Watt (P)
    where p = v x i i
    
I is the current strength in ampere units

V is a voltage in unit volts

```python
# Function Calculating Resistance
def hitungR(v,i):
    hasil = v/i
    return hasil

# The function of calculating power
def hitungP(v,i):
    hasil = v*i
    return hasil
```

Next, create a calculation system for test data, where the test conditions carried out are power/resistance with criteria for capacity> = 2.5 then do the power count, capacity < 2.5 then calculate the value of the resistance.

The results of the temporary calculation will be accommodated in the results column and will then be converted to column R/P as a result of the test calculation

```python
hasil = [] # Temporary value column
for i in range(len(data)):
    # Do the condition check on the battery capacity value
    if data["Cap"][i] < 2.5: #Untuk kondisi perhitungan hambatan
        hasil.append(hitungR(data["V_m"][i],data["Current (A)"][i]))
    elif data["Cap"][i] >= 2.5: #Untuk kondisi perhitungan daya
        hasil.append(hitungP(data["V_m"][i],data["Current (A)"][i]))
```

```python
# Create a new column with the name R/P which is the result of the value of the results column
data["R/P"] = hasil
```

After calculating according to the battery capacity condition and found the R/P column as a result.

Furthermore, calculations are made to find the average value of the R/P value with the condition of the capacity value for current = 1 ampere and odd cycle.

```python
# Displays capacity data (R/P) with the desired conditions
data.loc[(data["Current (A)"] == 1) & (data['Cycle'] % 2 != 0)]
```

```python
# Calculate with the desired conditions for finding the average value
dataBaru = data[(data['Current (A)'] == 1) & (data['Cycle'] % 2 != 0)]
rataRata = np.mean(dataBaru['Cap'])

print("The average value of the capacity value for current = 1 ampere and odd cycle is", rataRata)
```
